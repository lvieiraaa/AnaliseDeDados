```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")


dados_fraude = pd.DataFrame({
    'Ano': [2019, 2020, 2021, 2022, 2023],
    'Tipo_Fraude': [
        'Phishing',
        'Golpe do Telefone',
        'Fraude em Compras',
        'Estelionato',
        'Roubo'
    ],
    'Quantidade_Fraudes': [
        467361,  # Dados inspirados em relatórios do FBI
        317192,   # Dados inspirados em relatórios da FTC
        162000,   # Dados inspirados em relatórios de fraudes online
        2181500,  # Dados inspirados em relatórios sobre estelionato
        1300000   # Dados inspirados em relatórios sobre roubos
    ]
})


tabela_consolidada = dados_fraude.groupby(['Ano', 'Tipo_Fraude']).sum().reset_index()
tabela_consolidada = tabela_consolidada.rename(columns={"Quantidade_Fraudes": "Total de Fraudes"})

print("Tabela Consolidada:")
print(tabela_consolidada)



# Gráfico 1: Total de Fraudes por Ano
plt.figure(figsize=(10, 6))
sns.barplot(data=tabela_consolidada, x='Ano', y='Total de Fraudes', ci=None, estimator=sum)
plt.title('Total de Fraudes por Ano')
plt.ylabel('Total de Fraudes')
plt.xlabel('Ano')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('grafico_total_fraudes_por_ano.png')
plt.show()

# Gráfico 2: Fraudes por Tipo
plt.figure(figsize=(10, 6))
sns.barplot(data=tabela_consolidada, x='Tipo_Fraude', y='Total de Fraudes', ci=None, estimator=sum)
plt.title('Fraudes por Tipo')
plt.ylabel('Total de Fraudes')
plt.xlabel('Tipo de Fraude')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('grafico_fraudes_por_tipo.png')
plt.show()

# Gráfico 3: Fraudes por Ano e Tipo
plt.figure(figsize=(12, 7))
sns.lineplot(data=tabela_consolidada, x='Ano', y='Total de Fraudes', hue='Tipo_Fraude', marker='o')
plt.title('Total de Fraudes por Ano e Tipo')
plt.ylabel('Total de Fraudes')
plt.xlabel('Ano')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('grafico_fraudes_ano_tipo.png')
plt.show()

# 4. Salvar Tabela Consolidada em CSV
tabela_consolidada.to_csv('tabela_consolidada_fraudes.csv', index=False)
```
