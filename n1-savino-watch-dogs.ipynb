```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/kaggle/input/online-payment-fraud-detection/onlinefraud.csv'
dados = pd.read_csv(file_path)

print(dados.head())
```

```python
# Geração de uma tabela consolidada
tabela_consolidada = dados['isFraud'].value_counts()
print(tabela_consolidada)

# Gráfico de barras: Frequência de fraudes e transações normais
tabela_consolidada.plot(kind='bar', color=['blue', 'orange'])
plt.title('Distribuição de Fraudes e Transações Normais')
plt.xlabel('Tipo de Transação (0 = Normal, 1 = Fraude)')
plt.ylabel('Frequência')
plt.show()

# Gráfico de dispersão: Relação entre valor da transação e tipo de transação
plt.scatter(dados['amount'], dados['isFraud'], alpha=0.5)
plt.title('Relação entre Valor da Transação e Fraude')
plt.xlabel('Valor da Transação')
plt.ylabel('Fraude (0 = Não, 1 = Sim)')
plt.show()

# Gráfico de pizza: Proporção de fraudes
labels = ['Transações Normais', 'Fraudes']
plt.pie(tabela_consolidada, labels=labels, autopct='%1.1f%%', colors=['green', 'red'], startangle=90)
plt.title('Proporção de Fraudes')
plt.show()
```
