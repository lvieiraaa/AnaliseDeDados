```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/kaggle/input/online-payment-fraud-detection/onlinefraud.csv'
dados = pd.read_csv(file_path)

print(dados.head())
```

```python
# Gerao de uma tabela consolidada
tabela_consolidada = dados['isFraud'].value_counts()
print(tabela_consolidada)

# Grfico de barras: Frequncia de fraudes e transaes normais
tabela_consolidada.plot(kind='bar', color=['blue', 'orange'])
plt.title('Distribuio de Fraudes e Transaes Normais')
plt.xlabel('Tipo de Transao (0 = Normal, 1 = Fraude)')
plt.ylabel('Frequncia')
plt.show()

# Grfico de disperso: Relao entre valor da transao e tipo de transao
plt.scatter(dados['amount'], dados['isFraud'], alpha=0.5)
plt.title('Relao entre Valor da Transao e Fraude')
plt.xlabel('Valor da Transao')
plt.ylabel('Fraude (0 = No, 1 = Sim)')
plt.show()

# Grfico de pizza: Proporo de fraudes
labels = ['Transaes Normais', 'Fraudes']
plt.pie(tabela_consolidada, labels=labels, autopct='%1.1f%%', colors=['green', 'red'], startangle=90)
plt.title('Proporo de Fraudes')
plt.show()
```

```python
# Exemplo de como salvar um grfico
plt.savefig('grafico_fraude.png')
```
